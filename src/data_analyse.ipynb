```python
import pandas as pd 
import os 
```

```python
def read_images(path, label): 
 
    data = []
    folder_path = os.path.join(path, label)
    
    for i in os.listdir(folder_path): 
        if i.lower().endswith(('.png', '.jpg', '.jpeg')):

            img_path = os.path.join(folder_path, i) 
            data.append({'filepath': img_path, 'label': label})
        
        
    return pd.DataFrame(data)
```
